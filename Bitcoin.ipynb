```python
from scipy.stats import poisson
```

According to Mastering Bitcoin by O'Reilly "Every 10 minutes, on average, a Bitcoin miner can validate the transactions of the past 10 minutes and is rewarded with brand new bitcoin."
So for the h hours we would have h*60 minutes. 

```python
hours = int(input("Enter the number of hours for which you want to find how often the bitcoin network appear"))
```

```python
def calculate(hours,k):
  mu = int(hours*6)
  r = poisson.pmf(k,mu)
  return (1/r)/6

result = calculate(hours,k=0)
```

Using Poisson Distribuition, since It works well on independent events, giving probability of events occuring in a fixed time.
For which using the poisson function defined in Scipy As -  Screen Shot 2022-03-24 at 12.06.32 PM.png

Here the key arguments are k(number of times an event occur for the following interval) and mu(expected value of occurance), which can be calculated as  k = 0 since so mu = hours*60/10 min = (hours)x6

After which they can be passed to poissons pmf function, which will give a result in e^x which can be used to find the probability as 1/e^x * 10/60 hours

```python
def convert_hours_to_days(hours):
  days = hours//24
  return int(days)
days = convert_hours_to_days(result)
```
